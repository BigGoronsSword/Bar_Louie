```
import pandas as pd
import numpy as np
```

```
data = {'Shannon':[92.25,11.50,9,0], 'Dom':[10.99, 14,0,0], 'Grace':[20, 11,12,11]}
df = pd.DataFrame(data)
print(df)
```

```
#without patron
shannon_tip = df['Shannon'].sum() * 1.2
print(shannon_tip)
```

```
#without patron
dom_tip = df['Dom'].sum() * 1.2
print(dom_tip)
```

```
#without patron
grace_tip = df['Grace'].sum() * 1.2
print(grace_tip)
```

```
print(shannon_tip + dom_tip + grace_tip)
```

```
patron_tip_divided = (157.25 * 1.2) /4
print(patron_tip_divided)
```

```
with_patron_shannon = shannon_tip + patron_tip_divided
with_patron_dom = dom_tip + patron_tip_divided
with_patron_grace = grace_tip + patron_tip_divided
print('Shannon owes' , with_patron_shannon)
print('Dom owes', with_patron_dom)
print('Grace owes', with_patron_grace)
```
